```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =========================
# 0) Load data
# =========================
FILE_PATH = "outputs_by_model/gpt-5.2/n1/all_session_totals.csv"   # <-- change if needed
model_name = FILE_PATH.split("outputs_by_model/")[1].split("/")[0]
folder_name = FILE_PATH.split("outputs_by_model/")[1].split("/")[1]
print(f"Loading data for model: {model_name}, folder: {folder_name}")

df = pd.read_csv(FILE_PATH)

dims = [
    "species",
    "social_value",
    "gender",
    "age",
    "fitness",
    "utilitarianism",
    "interventionism",
    "relationship_to_vehicle",
    "concern_for_law",
]

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
display(df.head())

# check missing values
print("\nMissing values per column:")
print(df.isna().sum())

# Ensure expected columns exist
missing_dim_cols = [c for c in dims if c not in df.columns]
if missing_dim_cols:
    raise ValueError(f"Missing expected score columns in CSV: {missing_dim_cols}")


# =========================
# 1) Summary statistics
# =========================
summary = df[dims].describe().T
summary["range"] = summary["max"] - summary["min"]
display(summary)

# Mean + 95% CI
n = len(df)
means = df[dims].mean()
stds = df[dims].std(ddof=1)
se = stds / np.sqrt(n)
ci95_low = means - 1.96 * se
ci95_high = means + 1.96 * se

ci_table = pd.DataFrame({
    "mean": means,
    "std": stds,
    "ci95_low": ci95_low,
    "ci95_high": ci95_high,
}).sort_values("mean")

display(ci_table)


# =========================
# 2) Sign breakdown: positive / negative / zero rates
# =========================
sign_stats = []
for d in dims:
    sign_stats.append({
        "dimension": d,
        "pos_rate": (df[d] > 0).mean(),
        "neg_rate": (df[d] < 0).mean(),
        "zero_rate": (df[d] == 0).mean(),
        "mean": df[d].mean(),
        "median": df[d].median(),
        "mean_abs": df[d].abs().mean()
    })

sign_stats_df = pd.DataFrame(sign_stats).sort_values("mean")
display(sign_stats_df)


# =========================
# 3) Histograms (one plot per dimension)
# =========================
def plot_hist(series: pd.Series, title: str):
    plt.figure()
    series.hist(bins=15)
    plt.title(title)
    plt.xlabel("Score (per session)")
    plt.ylabel("Count")
    plt.show()

for col in dims:
    plot_hist(df[col], f"Histogram: {col}")


# =========================
# 4) Boxplot for 9 dimensions
# =========================
plt.figure()
df[dims].boxplot()
plt.title("Boxplot of 9 Dimension Scores (per session)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()


# =========================
# 5) Correlation heatmap
# =========================
corr = df[dims].corr()
display(corr)

plt.figure()
plt.imshow(corr, interpolation="nearest")
plt.title("Correlation Heatmap (9 Dimensions)")
plt.xticks(range(len(dims)), dims, rotation=45, ha="right")
plt.yticks(range(len(dims)), dims)
plt.colorbar()
plt.tight_layout()
plt.show()

# strongest correlations
corr_pairs = (
    corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
        .stack()
        .reset_index()
)
corr_pairs.columns = ["dim1", "dim2", "corr"]
corr_pairs["abs_corr"] = corr_pairs["corr"].abs()
display(corr_pairs.sort_values("abs_corr", ascending=False).head(15))


# =========================
# 6) Preference scale visualization (-1 to +1)
#    like your image but simplified (no others, no humans)
# =========================

# Normalization: map mean scores into [-1, +1]
# Core dims appear ~2 times per session -> max magnitude  2
# Interventionism is computed for every scenario -> max magnitude = 13
# Relationship_to_vehicle and concern_for_law could be <=13 (we normalize by 13 for a stable scale)
MAX_SCORE = {
    "species": 2,
    "social_value": 2,
    "gender": 2,
    "age": 2,
    "fitness": 2,
    "utilitarianism": 2,
    "interventionism": 13,
    "relationship_to_vehicle": 13,
    "concern_for_law": 13,
}

# Labels (what +1 means)
POS_END_LABEL = {
    "species": "Save humans",
    "social_value": "Save higher status",
    "gender": "Save female",
    "age": "Save younger",
    "fitness": "Save fitter",
    "utilitarianism": "Save more lives",
    "interventionism": "Intervene (swerve)",
    "relationship_to_vehicle": "Save passengers",
    "concern_for_law": "Save lawful",
}

NEG_END_LABEL = {
    "species": "Save animals",
    "social_value": "Save lower status",
    "gender": "Save male",
    "age": "Save older",
    "fitness": "Save less fit",
    "utilitarianism": "Save fewer lives",
    "interventionism": "Do not intervene (stay)",
    "relationship_to_vehicle": "Save pedestrians",
    "concern_for_law": "Save jaywalkers",
}

# normalized mean + CI
norm_mean = {}
norm_low = {}
norm_high = {}

for d in dims:
    m = means[d]
    low = ci95_low[d]
    high = ci95_high[d]
    denom = MAX_SCORE[d]

    norm_mean[d] = float(np.clip(m / denom, -1, 1))
    norm_low[d] = float(np.clip(low / denom, -1, 1))
    norm_high[d] = float(np.clip(high / denom, -1, 1))

# Create a stacked scale plot (single figure)
plt.figure(figsize=(12, 6))

y_positions = np.arange(len(dims))[::-1]  # top-to-bottom
for i, d in enumerate(dims):
    y = y_positions[i]

    # baseline scale line -1 to +1
    plt.plot([-1, 1], [y, y], linewidth=6, alpha=0.25)

    # center tick at 0
    plt.plot([0, 0], [y - 0.15, y + 0.15], linewidth=2)

    # 95% CI segment (thin)
    plt.plot([norm_low[d], norm_high[d]], [y, y], linewidth=3, alpha=0.8)

    # mean marker (vertical)
    plt.plot([norm_mean[d], norm_mean[d]], [y - 0.22, y + 0.22], linewidth=3)

    # left + right meaning labels (small)
    plt.text(-1.05, y, NEG_END_LABEL[d], va="center", ha="right", fontsize=9)
    plt.text(1.05, y, POS_END_LABEL[d], va="center", ha="left", fontsize=9)

    # dimension name
    plt.text(-0.02, y + 0.35, d, va="center", ha="center", fontsize=11, fontweight="bold")

plt.yticks([])
plt.xticks([-1, -0.5, 0, 0.5, 1])
plt.xlim(-1.2, 1.2)
plt.ylim(-1, len(dims))
plt.title(f"{model_name} Preference Scale by Dimension (normalized to -1 ~ +1)")
plt.xlabel("Preference Strength (negative  left meaning, positive  right meaning)")
plt.tight_layout()
plt.show()


# =========================
# 7) Save summary files (optional)
# =========================
summary.to_csv(f"outputs_by_model/{model_name}/{folder_name}/eda_summary_stats.csv")
ci_table.to_csv(f"outputs_by_model/{model_name}/{folder_name}/eda_means_ci95.csv")
corr.to_csv(f"outputs_by_model/{model_name}/{folder_name}/eda_correlations.csv")
sign_stats_df.to_csv(f"outputs_by_model/{model_name}/{folder_name}/eda_sign_stats.csv", index=False)

print(f"Saved: outputs_by_model/{model_name}/{folder_name}/eda_summary_stats.csv, outputs_by_model/{model_name}/{folder_name}/eda_means_ci95.csv, outputs_by_model/{model_name}/{folder_name}/eda_correlations.csv, outputs_by_model/{model_name}/{folder_name}/eda_sign_stats.csv")
```
